```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import pandas as pd
import time


driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
# 타겟 URL 설정
url = 'https://m.stock.naver.com/'  # 실제 크롤링할 페이지로 변경하세요
driver.get(url)
time.sleep(5)  # 페이지 로딩 대기

# 데이터가 있는 UL 태그 탐색
ul_tag = driver.find_element(By.XPATH, '//*[@class="LayoutCard_article__27nL4"]')  # XPATH는 실제 페이지에 맞게 수정해야 함
li_tags = ul_tag.find_elements(By.TAG_NAME, 'li')

# 데이터 추출
data = []
for li in li_tags:
    text = li.text
    if text:  # 텍스트가 있는 경우에만 추가
        data.append(text)

# 데이터프레임 생성
df = pd.DataFrame(data, columns=['Text'])

# 웹드라이버 종료
driver.quit()

# 결과 출력
print(df)
```

```python
##HTTP를 호출하는 프로그램을 작성할 때 주로 사용한다
import requests 
## html의 내용을 사용자가 활용하기 쉽게 하기위해 사용한다
from bs4 import BeautifulSoup 
##원하는 정보가 있는 주소
url = 'https://finance.naver.com/sise/sise_market_sum.naver?sosok=0' 
## 브라우저 호환을 위해서 설정
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 '+ \
    '(KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36'
            } 

##해당 url에 htlm(정보) 요청에사용 / url은 사용자가 원하는 url
r = requests.get(url, headers=headers, verify=True) 
## 해당 url의 html을 사용자가 활용하기 쉽게 변환
html = BeautifulSoup(r.text, 'html.parser')
##html에 주식정보가 있는 tag이며 find_all이기 때문에 리스트형태
tmp = html.find_all('tr', {'onmouseover': "mouseOver(this)"})
##리스트 형태이기 때문에 for문 사용 (for는 반복문 : 파이썬기초)
for i in tmp:
    ##주식정보가 있는 태그가 'td'이면서 list의 첫번째
    tmp2 = i.find_all('td')[1]
    ##주식명이 tag 'a'에 존재
    name = tmp2.find('a').text
    ##주식가격이 i의 tag 'td'의 list 3번째에 존재
    price = int(i.find_all('td')[2].text.replace(',',''))
    print([name,price])
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
import pymysql

# Selenium 옵션 설정
options = Options()
options.headless = False  # 브라우저를 열지 않고 실행 (디버깅을 위해 일단 False로 설정)

# 웹드라이버 설정
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=options)

def fetch_stock_data(driver):
    url = 'https://finance.naver.com/sise/sise_market_sum.naver?sosok=0'
    driver.get(url)
    
    # 페이지 로드 대기
    time.sleep(5)  # 동적 콘텐츠 로드를 위한 대기 시간

    # 주식정보가 있는 테이블 row 요소들 찾기
    rows = driver.find_elements(By.XPATH, '//table[@class="type_2"]/tbody/tr[not(@class="type03")][@onmouseover="mouseOver(this)"]')
    
    stock_data = []
    
    for row in rows:
        try:
            # 주식명이 있는 요소
            name_element = row.find_element(By.XPATH, './/td[2]/a')
            name = name_element.text

            # 주식가격이 있는 요소
            price_element = row.find_element(By.XPATH, './/td[3]')
            price = int(price_element.text.replace(',', ''))

            stock_data.append([name, price])
        except Exception as e:
            print(f"Error processing row: {e}")
            continue
    
    return stock_data

# 주식 데이터 크롤링
stock_data = fetch_stock_data(driver)

for data in stock_data:
    print(f"주식명: {data[0]}, 가격: {data[1]}")

driver.quit()

import csv  # csv 모듈 import 추가
# CSV 파일로 저장
csv_file = 'stock_data.csv'  # CSV 파일 경로
with open(csv_file, 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['주식명', '가격'])  # 헤더 쓰기
    for data in stock_data:
        writer.writerow(data)  # 각 주식 데이터 쓰기

print(f"주식 데이터가 {csv_file} 파일로 저장되었습니다.")
```

```python
import matplotlib.pyplot as plt

# 데이터 준비
stock_names = [data[0] for data in stock_data]
stock_prices = [data[1] for data in stock_data]

# 그래프 그리기
plt.figure(figsize=(7, 10))
plt.barh(stock_names, stock_prices, color='skyblue')
plt.xlabel('주식 가격')
plt.title('주식 가격 시각화')
# y축 눈금 레이블 기울이기
plt.yticks(rotation=20)

plt.tight_layout()
plt.show()
```
